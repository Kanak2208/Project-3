```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
happiness_indicators = pd.read_csv('/Users/ale/Desktop/MySQL/Project-3/world_happiness_report.csv')
economic_indicators = pd.read_csv('/Users/ale/Desktop/MySQL/Project-3/Economic Indicators And Inflation.csv')

# Quick check of the data
print(happiness_indicators.head())
print(economic_indicators.head())
```

```python
happiness_indicators.columns = (
    happiness_indicators.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('%', 'percent')
    .str.replace('(', '')
    .str.replace(')', '')
)
```

```python
economic_indicators.columns = (
    economic_indicators.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('%', 'percent')
    .str.replace('(', '')
    .str.replace(')', '')
)
```

```python
merged_data = pd.merge(
    happiness_indicators,
    economic_indicators,
    on=['country', 'year'],
    how='inner'
)
```

```python
merged_data.head()
```

```python
columns_to_drop = ['healthy_life_expectancy', 'freedom','unemployment_rate','unemployment_rate', 'employment_rate','mental_health_index','income_inequality','work_life_balance','inflation_rate_percent','political_stability','generosity', 'education_index', 'public_trust', 'public_health_expenditure','corruption_perception', 'climate_index', 'internet_access', 'crime_rate', 'gdp_in_billion_usd']
merged_data.drop(columns=columns_to_drop, inplace=True)
```

```python
grouped_data = merged_data.groupby(['country', 'year']).mean().round(2).reset_index()
grouped_data.head()
```
