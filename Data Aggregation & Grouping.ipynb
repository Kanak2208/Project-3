```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
happiness_indicators = pd.read_csv('/Users/ale/Desktop/MySQL/Project-3/world_happiness_report.csv')
economic_indicators = pd.read_csv('/Users/ale/Desktop/MySQL/Project-3/Economic Indicators And Inflation.csv')

# Quick check of the data
print(happiness_indicators.head())
print(economic_indicators.head())
```

```python
happiness_indicators.columns = (
    happiness_indicators.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('%', 'percent')
    .str.replace('(', '')
    .str.replace(')', '')
)
```

```python
economic_indicators.columns = (
    economic_indicators.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('%', 'percent')
    .str.replace('(', '')
    .str.replace(')', '')
)
```

```python
merged_data = pd.merge(
    happiness_indicators,
    economic_indicators,
    on=['country', 'year'],
    how='inner'
)
```
